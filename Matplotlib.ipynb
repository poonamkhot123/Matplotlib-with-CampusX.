```python
# colors (hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# line style dotted.
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted')

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# line style dashed.
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashed')

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# line style dashdot.
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot')

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# linewidth
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashed',linewidth=2)

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# marker D : Diamond
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',linewidth=3,marker = "D")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashed',linewidth=2,marker = "D")

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# marker +
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linewidth=1,marker="+")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linewidth=2,marker="+")

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# marker O
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linewidth=1,marker="o")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linewidth=2,marker="o")

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# markersize
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linewidth=1,marker="+",markersize="10")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linewidth=2,marker="+")

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')
```

```python
# legend-location
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linewidth=1,markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linewidth=2,markersize=8,label="Rohit")

plt.title('Rohit sharma Vs virat kohli career Comparison')
plt.xlabel('seasons')
plt.ylabel('Runs Scored')

plt.legend(loc='best')
```

```python
# color and label
df = pd.read_csv('/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
# width
plt.bar(df['batsman'],df['2015'],width = 0.5)
```

```python
# multiple bar charts
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='green')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='orange')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.show()
```

```python
# problem (if category name is very big)
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue blue','green green green green','yellow yellow yellow yellow','pink pink pink pink']
plt.bar(colors,children,color='blue')
plt.xticks(rotation='vertical')
```

**Stacked Bar chart**


```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']),label ='2015')
plt.legend()
plt.show()
```

**Pie chart**
* Univariate/Bivariate Analysis
* Categorical Vs Numerical
* Use case - To find contribution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
plt.pie(data)
plt.show()
```

```python
# data with labels.
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# dataset
df = pd.read_csv('/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
# what is the percentage of each section
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','green','yellow','pink','orange','grey'])
plt.show()
```

```python
# Explode Shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0])
plt.show()
```

```python
# explore shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```

```python
# different styles of plotting
plt.style.use('ggplot')
```

```python
# save figure
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)

plt.savefig('sample.png')
```
